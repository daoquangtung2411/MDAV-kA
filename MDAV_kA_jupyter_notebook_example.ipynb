## K-anonymity clustering using MDAV (Maximum Distance to Average Vector) microaggregation techniques

```python
import pandas as pd
import matplotlib.pyplot as plt
from MDAV_kA import anonymized_dataset, cal_info_loss
```

```python
dataset = pd.read_csv('health_ai_mdav_demo.csv')
k_list = list(range(3,13,1))
sse_list = []
origin_dataset = dataset.copy()
for k in k_list:
    anonymized_df = anonymized_dataset(dataset, k=k)
    sse = cal_info_loss(origin_dataset, anonymized_df)
    sse_list.append(sse)
plt.plot(k_list, sse_list)
plt.xlabel('k-anonymity')
plt.xticks(k_list)
plt.show()
```
